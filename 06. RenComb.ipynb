# Renombrado y Combinación

En este JNB se revisará el tema de Renombrado y Combinación en Pandas.

```python
import pandas as pd
reviews = pd.read_csv("https://raw.githubusercontent.com/GTD2101-ACD-AA-EJ23/pandas-20170778/main/data/winemag-data-130k-v2.csv", index_col=0)
pd.set_option("display.max_rows", 5)
```

```python
reviews.rename(columns={'points': 'score'})
```

```python
reviews # no cambia 
```

```python
reviews.rename(index={0: 'firstEntry', 1: 'secondEntry'})
```

```python
reviews.index
```

```python
reviews.columns
```

```python
reviews.rename_axis("wines", axis="rows").rename_axis("fields", axis="columns")
```
